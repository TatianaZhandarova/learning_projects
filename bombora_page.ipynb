```python
# скрейпинг
# подключение библиотек:
import requests #используется для отправки запросов на внешние ресурсы
from bs4 import BeautifulSoup
```

```python
# задание адреса для отправки запроса
url = 'https://bombora.ru/books/?bestseller=da'
#задание адреса для каждой страницы, меняем только page-{namber_page}
#address_by_page = "https://bombora.ru/books/page-4/?bestseller=da"
```

```python
# отправка запроса по указанному адресу
page = requests.get(url)
```

```python
# анализ ответа
type(page)
page.status_code

# текст ответа
#page.text
```

```python
# с помощью юиюилиотеки bs4 и BeautifulSoup
# конструктором из нее делаем разметку полученного текста
soup = BeautifulSoup(page.text, 'lxml')
```

```python
#  разбор элементов переменной soup
list_results = soup.find_all('div', attrs={'class': ['book-card2', 'book-card2', 'book-card2', 'book-card2']})
```

```python
#проверка, правильно ли все вписала
len(list_results)
list_results[0]
```

```python
#разбор данных
list_results[0].find('a', attrs={'class': 'book-card2__title'}).get_text()
list_results[0].find('div', attrs={'class': 'book-card2__author'}).get_text()
list_results[0].find('div', attrs={'class': 'book-card2__price'}).get_text(strip=True)
internet_page = 'https://bombora.ru/' + list_results[0].find('a').get('href')
print(internet_page)
```

```python
# подключение библиотек для дальнешего оформления данных
import pandas as pd
import numpy as np
```

```python
#формирование списков
list_books_name = []
for item in list_results:
  list_1 = list(item.find('a', attrs={'class': 'book-card2__title'}).get_text().split(" "))
  list_books_name.append(list_1)
print(list_books_name)
```

```python
list_autor_name = []
for item in list_results:
  list_2 = list(item.find('div', attrs={'class': 'book-card2__author'}).get_text().split(" "))
  list_autor_name.append(list_2)
print(list_autor_name)
```

```python
list_prise = []
for item in list_results:
  list_3 = list(item.find('div', attrs={'class': 'book-card2__price'}).get_text(strip=True).split(","))
  list_prise.append(list_3)
#print(list_prise)

str_prise = str(list_prise).replace(" ", "")
print(str_prise)
```

```python
list_rel =[]
for item in list_results:
  list_4 = list(('https://bombora.ru/' +  item.find('a').get('href')).split(","))
  list_rel.append(list_4)
print(list_rel)
```

```python
books_data = {'Название книги' : [item[0] for item in list_books_name],
               'Автор' : [item[0] for item in list_autor_name],
               'Цена на печатную и электронную верисю' : [item[0] for item in list_prise],
               'Ссылка на выбранную книгу' : [item[0] for item in list_rel]
               }
print(books_data)
pd.DataFrame(books_data)
```
